```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
data= pd.read_csv("/content/homeprices.csv")
```

```python
data
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
data.corr()
```

```python
p=data.drop("price",axis=1)
p
```

```python
q=data["price"]
q
```

```python
print("p shape is : ",p.shape)
print("q shape is : ",q.shape)
```

```python
plt.scatter(p,q)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(p,q,test_size=0.30,shuffle=True,random_state=40)
print("x shape is : ",x_train.shape)
print("x shape is : ",x_test.shape)
print("y shape is : ",y_train.shape)
print("y shape is : ",x_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
plt.plot(x_train,y_train,'bo')
```

```python
plt.plot(x_test,y_test,'ro')
```

```python
y_pred=model.predict(x_test)
plt.plot(y_pred,y_test,'ro')
plt.plot(y_test,y_test,'b-')
```

```python
model.score(x_test,y_test)
```

```python
model.predict([[3300]])
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
t=model.coef_
```

```python
u=model.intercept_
```

```python
x=t*3300+u
```
